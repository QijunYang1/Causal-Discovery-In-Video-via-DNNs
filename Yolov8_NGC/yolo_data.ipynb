## Change the bbox into center points format and normalized the results

```python
if not os.path.isdir("../data/custom/labels"):
    os.mkdir("../data/custom/labels")
def bbox(frame, df):
    file = "../data/custom/labels/{:06d}.txt".format(int(frame))
    f = open(file, 'w')
    for _, row in df.iterrows():
        id = int(row['id'])
        width = float(row['bb_width'])
        height = float(row['bb_height'])
        x_center = float(row['bb_left']) + 0.5 * width
        y_center = float(row['bb_top']) + 0.5 * height
        width /= 1280
        x_center/=1280
        y_center/=720
        height/=720
        f.write(" ".join(str(e) for e in [id, x_center,y_center,width,height]) + "\n")

        
    f.close()
for key, df in gt.groupby('frame'):
    
    bbox(key, df)
```
